```python
from sklearn.feature_extraction import DictVectorizer
import csv
from sklearn import tree
from sklearn import preprocessing
from six import StringIO
import json
import pandas as pd
import os
from prettytable import PrettyTable
```

```python
def read_csv_to_list(path):
    messages = []
    
    for root,dirs,files in os.walk(path):
        for dir in dirs:
            for r,ds,fs in os.walk(dir):
                for f in fs:
                    readFile(messages, r, f)
        for file in files:
            readFile(messages, root, file)

    return messages
```

```python
def getAddrName(str):
    return str.split('_txs.json')[0]
```

```python
def readFile(messages, rootName, fileName):
    try:
        addrName = getAddrName(fileName)
        f = open(os.path.join(rootName, fileName))

        line = f.readline()
        while line:
            transaction_json = json.loads(line)
            messages.append(transaction_json)
            line = f.readline()
    except Exception as ex:
        print(fileName)

    f.close()
```

```python
# trans = {一系列交易}
# addr_trans = {某一个地址: [ [相关事务id], [传入交易时间], [传出交易时间], 传入数量, 传入金额, 传出数量, 传出金额, 
#                          存在自己传给自己的现象的事务数量, 自己传给自己的现象的事务的金额总数]
#               第二个地址: [...]}

#[[事务id], [传入数量， 传入金额, 传入交易时间], [传出数量, 传出金额, 传出交易时间], [自传自数量, 金额, 时间]]
```

```python
trans = read_csv_to_list(path)
addr_trans = {}

trash_pool = []

for tran in trans:
    if tran in trash_pool:
        continue
    trash_pool.append(tran)
    
    in_addr = []
    out_addr = []
    
    for vin in tran['vin']:
        if 'address' not in vin:
            continue
            
        addr = vin['address']
        if addr not in addr_trans:
            addr_trans[addr] = [[],[],[],0,0, 0,0, 0,0]            

        
        addr_trans[addr][0].append(tran['_id'])  
        addr_trans[addr][3] += 1
        addr_trans[addr][4] += vin['value']
        if addr not in in_addr:
            in_addr.append(addr)
        
    for vout in tran['vout']:
        addr = vout['address']
        if addr not in addr_trans:
            addr_trans[addr] = [[],[],[],0,0, 0,0, 0,0]            

        
        addr_trans[addr][0].append(tran['_id'])
        addr_trans[addr][5] += 1
        addr_trans[addr][6] += vout['value']
        
        if addr not in out_addr:
            out_addr.append(addr)
            
        if addr in in_addr:
            addr_trans[addr][7] += 1
            addr_trans[addr][8] += vout['value']
            
    
    for addr in in_addr:
        addr_trans[addr][1].append(tran['timestamp'])
    for addr in out_addr:
        addr_trans[addr][2].append(tran['timestamp'])
      
addr_trans
```
